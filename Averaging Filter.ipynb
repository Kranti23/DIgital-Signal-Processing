```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(1,20)
pure=np.sin(x)
plt.subplot(3,1,1)
plt.plot(x,pure)
plt.title('Purewave')

noise=np.random.normal(scale=0.1,size=len(x))

y=pure+noise
plt.subplot(3,1,2)
plt.plot(x,y)
plt.title('Distortedwave')

win_size=3
for i in range(1,(len(y)-1)):
    y[i]=(y[i-1]+y[i]+y[i+1]/win_size)

plt.subplot(3,1,3)
plt.plot(x,y)
plt.title("AveragingFilter")
plt.show()
```
